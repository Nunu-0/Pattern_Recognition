### [과제#1] 영상 히스토그램 그리기
임의의 컬러 영상 하나를 읽어서 grayscale로 변환 후, 히스토그램을 그리는 코드를 작성하세요.

※ cv2 및 matplotlib 라이브러리를 활용한 히스토그램 생성 방법을 스스로 찾아보고 스터디 후 구현하면 됩니다.

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

# 이미지 회색화
img_gray = cv2.imread('Hedgehog.jpg', cv2.IMREAD_GRAYSCALE)

# 회색화 한 이미지 보이기
plt.subplot(2, 1, 1)
plt.imshow(img_gray, cmap = 'gray')

# 이미지 히스토그램 보이기
plt.subplot(2, 1, 2)
plt.hist(img_gray.ravel(), 256, [0,256])

plt.show()
```
